```python
import re

# Step 1: Read the YOLO output from the file
with open("yolo_output.txt", "r") as file:
    yolo_output = file.read()


# Step 1: Define irrelevant words to remove
irrelevant_words = ["torch", "CUDA", "layers", "parameters", "gradients", "GFLOPs" , "1"]

# Step 2: Initialize a parent list to store lists of objects for each image
all_images_objects = []

# Step 3: Split the output by 'image' to process each image's section separately
image_sections = yolo_output.split("image ")[1:]  # Splits output by each image, ignoring the first part

# Step 4: Process each image section
for section in image_sections:
    # Extract detected objects (numbers followed by words)
    detected_objects = re.findall(r"\d+ (\w+)", section)

    # Filter out irrelevant objects
    filtered_objects = [obj for obj in detected_objects if obj not in irrelevant_words]

    # Append each image's objects as a separate list
    all_images_objects.append(filtered_objects)

# Step 5: Output the list of lists
for idx, objects_list in enumerate(all_images_objects):
    print(f"Image {idx+1} detected objects: {objects_list}")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from huggingface_hub import login

login(token="Add Your Token Here")

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the LLAMA 3.1 model and tokenizer with 4-bit quantization
llama_model_path = "meta-llama/Meta-Llama-3-8B-Instruct"
llama_tokenizer = AutoTokenizer.from_pretrained(llama_model_path)
llama_model = AutoModelForCausalLM.from_pretrained(
    llama_model_path,
    quantization_config=quantization_config
)
```

```python
llama_model.to(device)

# Function to generate a recipe prompt for each image
def create_recipe_prompt(ingredients):
    ingredients_str = ", ".join(ingredients)
    return f"Create a recipe using the following ingredients: {ingredients_str}."

# Function to generate a response from the model
def generate_response(prompt, max_length=400):
    inputs = llama_tokenizer(prompt, return_tensors="pt").to(device)
    with torch.no_grad():
        output = llama_model.generate(
            inputs.input_ids, 
            max_length=max_length, 
            do_sample=True,
            top_k=50,
            top_p=0.95,
            temperature=0.7
        )
    return llama_tokenizer.decode(output[0], skip_special_tokens=True)

# Step 11: Generate and print recipes for each image's detected objects
for idx, objects_list in enumerate(all_images_objects):
    prompt = create_recipe_prompt(objects_list)
    recipe = generate_response(prompt)
    print(f"Recipe for Image {idx+1}:")
    print(recipe)
    print("-" * 50)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

# Custom Dataset class for tokenized inputs
class TextDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __len__(self):
        return len(self.encodings['input_ids'])

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        return item

# Create Dataset objects
train_dataset = TextDataset(train_encodings)
val_dataset = TextDataset(val_encodings)

# Create DataLoader objects
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=8)
```
